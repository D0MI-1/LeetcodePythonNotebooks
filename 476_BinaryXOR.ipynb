476. Number Complement

The complement of an integer is the integer you get when you flip all the 0's to 1's and all the 1's to 0's in its binary representation.

For example, The integer 5 is "101" in binary and its complement is "010" which is the integer 2.
Given an integer num, return its complement.



Example 1:

Input: num = 5
Output: 2
Explanation: The binary representation of 5 is 101 (no leading zero bits), and its complement is 010. So you need to output 2.
Example 2:

Input: num = 1
Output: 0
Explanation: The binary representation of 1 is 1 (no leading zero bits), and its complement is 0. So you need to output 0.

```python
class Solution:
    def findComplement(self, num: int) -> int:
        result = []

        binary = bin(num)  # Get binary representation of `num`
        binary_without_prefix = binary[2:]  # Remove "0b" prefix
        for char in binary_without_prefix:  # Loop through each binary digit
            if char == '1':
                result.append('0')  # Replace '1' with '0'
            else:
                result.append('1')  # Replace '0' with '1'

        result = "".join(result)  # Join the result list into a string

        return int(result, 2)  # Convert the binary string back to an integer
```

```python
class Solution:
    def findComplement(self, num: int) -> int:
        bit_length = num.bit_length()  # Get the number of bits needed to represent `num`

        mask = (1 << bit_length) - 1  # Create a mask with all bits set to 1 (same length as `num`)

        return num ^ mask  # XOR the number with the mask to flip all its bits
```

bit_length: num = 5 returns 3 (101)


---


(1 << bit_length) - 1:
  
  * 1 << bit_length creates a binary number with one bit followed by bit_length zeros
  
  * e.g. bit_length 3 gives 1000 in binary
  subtracting 1 from 1000 gives us 111 this mask is used to flip all bits of num


---


num ^ mask: XOR (^) flips the bits

*   e.g.: num 5 (101)

*   num : mask

*     1 : 1 = 0

*     0 : 1 = 1

*     1 : 1 = 0



```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

def find_complement(num):
    bit_length = num.bit_length()
    mask = (1 << bit_length) - 1
    complement = num ^ mask

    states = [
        {'step': 'Original Number', 'value': num, 'binary': f'{num:b}'.zfill(bit_length)},
        {'step': 'Mask', 'value': mask, 'binary': f'{mask:b}'.zfill(bit_length)},
        {'step': 'Complement', 'value': complement, 'binary': f'{complement:b}'.zfill(bit_length)}
    ]

    return complement, states

def visualize_complement(num):
    result, states = find_complement(num)

    fig, ax = plt.subplots(figsize=(10, 6))
    fig.suptitle(f'Number Complement Visualization (num={num})', fontsize=16)

    def update(frame):
        ax.clear()
        state = states[frame]

        # Plot binary representation
        for i, bit in enumerate(state['binary']):
            color = 'red' if bit == '1' else 'blue'
            ax.text(i, 0, bit, color=color, fontsize=20, ha='center', va='center')

        ax.set_xlim(-1, len(state['binary']))
        ax.set_ylim(-1, 1)
        ax.axis('off')

        # Add text annotations
        ax.text(0.02, 0.95, f"Step: {state['step']}", transform=ax.transAxes, verticalalignment='top', fontsize=12)
        ax.text(0.02, 0.90, f"Value: {state['value']}", transform=ax.transAxes, verticalalignment='top', fontsize=12)
        ax.text(0.02, 0.85, f"Binary: {state['binary']}", transform=ax.transAxes, verticalalignment='top', fontsize=12)

    anim = FuncAnimation(fig, update, frames=len(states), interval=2000, repeat=False)
    plt.close(fig)
    return anim, result

# Example usage
num = 5
anim, complement_result = visualize_complement(num)

# Display the animation in Colab

print(f"The complement of {num} is: {complement_result}")

HTML(anim.to_jshtml())
```
