145. Binary Tree Postorder Traversal

Given the root of a binary tree, return the postorder traversal of its nodes' values.



Example 1:


Input: root = [1,null,2,3]
Output: [3,2,1]
Example 2:

Input: root = []
Output: []
Example 3:

Input: root = [1]
Output: [1]


Constraints:

The number of the nodes in the tree is in the range [0, 100].
-100 <= Node.val <= 100


```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def postorderTraversal(self, root: Optional[TreeNode]) -> List[int]:

        result = []

        def traverse(node):
            if node is None:
                return

            traverse(node.left)
            traverse(node.right)
            result.append(node.val)

        traverse(root)
        return result
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def postorder_traversal(root):
    result = []
    states = []

    def traverse(node, depth=0):
        if node is None:
            return

        states.append({'node': node.val, 'depth': depth, 'action': 'enter'})
        traverse(node.left, depth + 1)
        traverse(node.right, depth + 1)
        result.append(node.val)
        states.append({'node': node.val, 'depth': depth, 'action': 'exit'})

    traverse(root)
    return result, states

def visualize_postorder_traversal(root):
    result, states = postorder_traversal(root)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    fig.suptitle('Postorder Traversal Visualization', fontsize=16)

    def plot_tree(ax, root, x=0, y=0, dx=1, dy=-1):
        if root is None:
            return
        ax.plot(x, y, 'o', markersize=20, color='lightblue')
        ax.text(x, y, str(root.val), ha='center', va='center')
        if root.left:
            ax.plot([x, x-dx], [y, y+dy], '-k')
            plot_tree(ax, root.left, x-dx, y+dy, dx/2, dy)
        if root.right:
            ax.plot([x, x+dx], [y, y+dy], '-k')
            plot_tree(ax, root.right, x+dx, y+dy, dx/2, dy)

    plot_tree(ax1, root)
    ax1.axis('off')
    ax1.set_title('Binary Tree')

    visited = set()
    result_text = ax2.text(0.1, 0.9, '', fontsize=12, verticalalignment='top', wrap=True)

    def update(frame):
        state = states[frame]
        node = state['node']
        depth = state['depth']
        action = state['action']

        if action == 'enter':
            color = 'yellow'
        else:  # exit
            color = 'lightgreen'
            visited.add(node)

        for i, (x, y, val) in enumerate(ax1.findobj(match=plt.Circle)):
            if val.get_text() == str(node):
                val.set_color(color)
                break

        result_text.set_text('Traversal order: ' + ' '.join(map(str, [n for n in visited])))
        ax2.set_title(f"Step {frame+1}/{len(states)}: {'Visiting' if action == 'enter' else 'Finished'} node {node}")
        ax2.axis('off')

    anim = FuncAnimation(fig, update, frames=len(states), interval=1000, repeat=False)
    plt.close(fig)
    return anim, result

# Example usage
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

anim, traversal_result = visualize_postorder_traversal(root)

# Display the animation in Colab

print(f"The postorder traversal result is: {traversal_result}")

HTML(anim.to_jshtml())
```
