947. Most Stones Removed with Same Row or Column

On a 2D plane, we place n stones at some integer coordinate points. Each coordinate point may have at most one stone.

A stone can be removed if it shares either the same row or the same column as another stone that has not been removed.

Given an array stones of length n where stones[i] = [xi, yi] represents the location of the ith stone, return the largest possible number of stones that can be removed.



Example 1:

Input: stones = [[0,0],[0,1],[1,0],[1,2],[2,1],[2,2]]
Output: 5
Explanation: One way to remove 5 stones is as follows:
1. Remove stone [2,2] because it shares the same row as [2,1].
2. Remove stone [2,1] because it shares the same column as [0,1].
3. Remove stone [1,2] because it shares the same row as [1,0].
4. Remove stone [1,0] because it shares the same column as [0,0].
5. Remove stone [0,1] because it shares the same row as [0,0].
Stone [0,0] cannot be removed since it does not share a row/column with another stone still on the plane.
Example 2:

Input: stones = [[0,0],[0,2],[1,1],[2,0],[2,2]]
Output: 3
Explanation: One way to make 3 moves is as follows:
1. Remove stone [2,2] because it shares the same row as [2,0].
2. Remove stone [2,0] because it shares the same column as [0,0].
3. Remove stone [0,2] because it shares the same row as [0,0].
Stones [0,0] and [1,1] cannot be removed since they do not share a row/column with another stone still on the plane.
Example 3:

Input: stones = [[0,0]]
Output: 0
Explanation: [0,0] is the only stone on the plane, so you cannot remove it.

```python
class UnionFind:
    def __init__(self):
        self.parent = {}
        self.rank = {}

    def find(self, x):
        if x not in self.parent:
            self.parent[x] = x
            self.rank[x] = 0
            return x
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        px, py = self.find(x), self.find(y)
        if px == py:
            return
        if self.rank[px] < self.rank[py]:
            self.parent[px] = py
        elif self.rank[px] > self.rank[py]:
            self.parent[py] = px
        else:
            self.parent[py] = px
            self.rank[px] += 1

class Solution:
    def removeStones(self, stones):
        uf = UnionFind()
        for x, y in stones:
            uf.union(x, ~y)  # Use ~y to distinguish y coordinates from x coordinates

        components = set(uf.find(x) for x, _ in stones) | set(uf.find(~y) for _, y in stones)
        return len(stones) - len(components)

# Example usage
stones = [[0,0],[0,1],[1,0],[1,2],[2,1],[2,2]]
solution = Solution()
result = solution.removeStones(stones)
print(f"Input: {stones}")
print(f"Number of stones that can be removed: {result}")

# Additional example
stones2 = [[0,0],[0,2],[1,1],[2,0],[2,2]]
result2 = solution.removeStones(stones2)
print(f"\nInput: {stones2}")
print(f"Number of stones that can be removed: {result2}")
```

```python
class UnionFind:
    def __init__(self):
        self.parent = {}
        self.rank = {}
        print("Initialized empty UnionFind with parent: {}, rank: {}".format(self.parent, self.rank))

    def find(self, x):
        print(f"\nFinding root for {x}")
        if x not in self.parent:
            self.parent[x] = x
            self.rank[x] = 0
            print(f"New element {x} added. parent[{x}] = {x}, rank[{x}] = 0")
            return x
        if self.parent[x] != x:
            print(f"{x} is not its own parent. Current parent[{x}] = {self.parent[x]}")
            self.parent[x] = self.find(self.parent[x])
            print(f"After path compression, parent[{x}] = {self.parent[x]}")
        print(f"Returning root {self.parent[x]} for {x}")
        return self.parent[x]

    def union(self, x, y):
        print(f"\nUnioning {x} and {y}")
        px, py = self.find(x), self.find(y)
        print(f"Root of {x} is {px}, root of {y} is {py}")
        if px == py:
            print(f"{x} and {y} are already in the same set")
            return
        if self.rank[px] < self.rank[py]:
            self.parent[px] = py
            print(f"Rank of {px} < {py}, so parent[{px}] = {py}")
        elif self.rank[px] > self.rank[py]:
            self.parent[py] = px
            print(f"Rank of {px} > {py}, so parent[{py}] = {px}")
        else:
            self.parent[py] = px
            self.rank[px] += 1
            print(f"Ranks equal, so parent[{py}] = {px} and rank[{px}] increased to {self.rank[px]}")
        print(f"After union: parent = {self.parent}, rank = {self.rank}")

def removeStones(stones):
    uf = UnionFind()
    for i, (x, y) in enumerate(stones):
        print(f"\nProcessing stone {i}: ({x}, {y})")
        uf.union(x, ~y)

    components = set(uf.find(x) for x, _ in stones) | set(uf.find(~y) for _, y in stones)
    print(f"\nFinal parent: {uf.parent}")
    print(f"Final rank: {uf.rank}")
    print(f"Unique components: {components}")
    return len(stones) - len(components)

# Example usage
stones = [[0,0],[0,1],[1,0],[1,2],[2,1],[2,2]]
result = removeStones(stones)
print(f"\nInput: {stones}")
print(f"Number of stones that can be removed: {result}")
```
