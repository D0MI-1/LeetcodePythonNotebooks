650. 2 Keys Keyboard

There is only one character 'A' on the screen of a notepad. You can perform one of two operations on this notepad for each step:

Copy All: You can copy all the characters present on the screen (a partial copy is not allowed).
Paste: You can paste the characters which are copied last time.
Given an integer n, return the minimum number of operations to get the character 'A' exactly n times on the screen.



Example 1:

Input: n = 3
Output: 3
Explanation: Initially, we have one character 'A'.
In step 1, we use Copy All operation.
In step 2, we use Paste operation to get 'AA'.
In step 3, we use Paste operation to get 'AAA'.
Example 2:

Input: n = 1
Output: 0

```python
class Solution:
    def minSteps(self, n: int) -> int:
        solution = 0

        if n == 1:
            return 0


        while n % 2 == 0:
            solution += 2
            n //= 2

        for i in range(3, int(n**0.5) + 1, 2):
            while n % i == 0:
                solution += i
                n //= i

        if n > 2:
            solution += n


        return solution
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

def min_steps(n):
    solution = 0
    states = []
    divisors = []
    current_n = n

    if n == 1:
        return 0, states, divisors

    while current_n % 2 == 0:
        solution += 2
        divisors.append(2)
        current_n //= 2
        states.append({'divisor': 2, 'n_before': current_n * 2, 'n_after': current_n, 'solution': solution, 'divisors': divisors.copy()})

    for i in range(3, int(current_n**0.5) + 1, 2):
        while current_n % i == 0:
            solution += i
            divisors.append(i)
            current_n //= i
            states.append({'divisor': i, 'n_before': current_n * i, 'n_after': current_n, 'solution': solution, 'divisors': divisors.copy()})

    if current_n > 2:
        solution += current_n
        divisors.append(current_n)
        states.append({'divisor': current_n, 'n_before': current_n, 'n_after': 1, 'solution': solution, 'divisors': divisors.copy()})

    return solution, states, divisors

def visualize_min_steps(n):
    result, states, divisors = min_steps(n)

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
    fig.suptitle(f'Minimum Steps Visualization (n={n})', fontsize=16)

    def update(frame):
        ax1.clear()
        ax2.clear()

        state = states[frame]

        # Plot division process
        ax1.text(0.5, 0.6, f"{state['n_before']} รท {state['divisor']} = {state['n_after']}",
                 horizontalalignment='center', verticalalignment='center', fontsize=20)
        ax1.axis('off')
        ax1.set_title(f'Division Process (Step {frame+1}/{len(states)})')

        # Plot divisors array
        ax2.bar(range(len(state['divisors'])), state['divisors'])
        ax2.set_title('Divisors Array')
        ax2.set_xlabel('Index')
        ax2.set_ylabel('Divisor')

        # Add text annotations
        ax2.text(0.02, 0.95, f"Current n: {state['n_after']}", transform=ax2.transAxes, verticalalignment='top')
        ax2.text(0.02, 0.90, f"Current Solution: {state['solution']}", transform=ax2.transAxes, verticalalignment='top')
        ax2.text(0.02, 0.85, f"Divisors: {state['divisors']}", transform=ax2.transAxes, verticalalignment='top')

    anim = FuncAnimation(fig, update, frames=len(states), interval=1500, repeat=False)
    plt.close(fig)
    return anim, result, divisors

# Example usage
n = 18
anim, min_steps_result, final_divisors = visualize_min_steps(n)

# Display the animation in Colab

print(f"The minimum number of steps for n={n} is: {min_steps_result}")
print(f"The divisors are: {final_divisors}")

HTML(anim.to_jshtml())
```
