264. Ugly Number II

An ugly number is a positive integer whose prime factors are limited to 2, 3, and 5.

Given an integer n, return the nth ugly number.



Example 1:

Input: n = 10
Output: 12
Explanation: [1, 2, 3, 4, 5, 6, 8, 9, 10, 12] is the sequence of the first 10 ugly numbers.
Example 2:

Input: n = 1
Output: 1
Explanation: 1 has no prime factors, therefore all of its prime factors are limited to 2, 3, and 5.

```python
class Solution:
    def nthUglyNumber(self, n: int) -> int:
        ugly = [1]
        i2, i3, i5 = 0, 0, 0

        while len(ugly) < n:
            u2, u3, u5 = ugly[i2] * 2, ugly[i3] * 3, ugly[i5] * 5
            ugly.append(min(u2, u3, u5))

            if ugly[-1] == u2:
                i2 += 1
            if ugly[-1] == u3:
                i3 += 1
            if ugly[-1] == u5:
                i5 += 1

        return ugly[-1]
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
from IPython.display import HTML

def nth_ugly_number(n):
    ugly = [1]
    i2, i3, i5 = 0, 0, 0
    states = []

    while len(ugly) < n:
        u2, u3, u5 = ugly[i2] * 2, ugly[i3] * 3, ugly[i5] * 5
        next_ugly = min(u2, u3, u5)
        ugly.append(next_ugly)

        states.append({
            'ugly': ugly.copy(),
            'i2': i2,
            'i3': i3,
            'i5': i5,
            'u2': u2,
            'u3': u3,
            'u5': u5,
            'next_ugly': next_ugly
        })

        if next_ugly == u2:
            i2 += 1
        if next_ugly == u3:
            i3 += 1
        if next_ugly == u5:
            i5 += 1

    return ugly[-1], states

def visualize_nth_ugly_number(n):
    result, states = nth_ugly_number(n)

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
    fig.suptitle(f'Nth Ugly Number Visualization (n={n})', fontsize=16)

    def update(frame):
        ax1.clear()
        ax2.clear()

        state = states[frame]

        # Plot ugly numbers
        ax1.plot(state['ugly'], 'bo-')
        ax1.set_title(f'Ugly Numbers (Step {frame+1}/{len(states)})')
        ax1.set_xlabel('Index')
        ax1.set_ylabel('Ugly Number')

        # Highlight the latest ugly number
        ax1.plot(len(state['ugly'])-1, state['ugly'][-1], 'ro', markersize=10)

        # Plot pointers and next possible ugly numbers
        x = ['i2', 'i3', 'i5']
        y = [state['u2'], state['u3'], state['u5']]
        ax2.bar(x, y)
        ax2.set_title('Next Possible Ugly Numbers')
        ax2.set_ylabel('Value')

        # Highlight the chosen next ugly number
        chosen_index = x[y.index(state['next_ugly'])]
        ax2.bar(chosen_index, state['next_ugly'], color='red')

        # Add text annotations
        ax2.text(0.02, 0.95, f"i2: {state['i2']}", transform=ax2.transAxes, verticalalignment='top')
        ax2.text(0.02, 0.90, f"i3: {state['i3']}", transform=ax2.transAxes, verticalalignment='top')
        ax2.text(0.02, 0.85, f"i5: {state['i5']}", transform=ax2.transAxes, verticalalignment='top')
        ax2.text(0.02, 0.80, f"Next Ugly: {state['next_ugly']}", transform=ax2.transAxes, verticalalignment='top')

    anim = FuncAnimation(fig, update, frames=len(states), interval=1000, repeat=False)
    plt.close(fig)
    return anim, result

# Example usage
n = 15
anim, nth_ugly = visualize_nth_ugly_number(n)

# Display the animation in Colab

print(f"The {n}th ugly number is: {nth_ugly}")

HTML(anim.to_jshtml())
```
